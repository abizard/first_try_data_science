```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset Freepass DS 2023.csv', delimiter=';')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

# Data Cleaning

```python
for col in df.columns :
    df[col] = df[col].str.replace('\r','').str.replace('\n','')
```

```python
df.head()
```

**Rename Columns Header**

```python
df_new = df.copy()
```

```python
df_new.rename(columns={'judul':'Title', 'link':'Link', 'date':'Publication_date', 'author':'Author', 'abstract':'Abstract'}, inplace=True)
```

```python
df_new.head(10)
```

**Cleaning Publication_date Columns**

```python
df_new['Publication_date'] = df_new['Publication_date'].str.replace('Diterbitkan','').str.replace('\t','')
```

```python
df_new.head()
```

```python
df_new['Publication_date'] = (df_new['Publication_date'].str.replace('Peb','Feb')
														.str.replace('Mei','May')
														.str.replace('Agu','Aug')
														.str.replace('Okt','Oct')
														.str.replace('Des','Dec'))
df_new.head(20)
```

```python
df_new['Publication_date'] = pd.to_datetime(df_new['Publication_date'])
```

```python
df_new.head()
```

**Add Month and Year Column**

```python
df_new['Month'] = pd.DatetimeIndex(df_new['Publication_date']).month
df_new['Year'] = pd.DatetimeIndex(df_new['Publication_date']).year
```

```python
df_new.head()
```

**Cleaning Title Column**

```python
df_new['Title'] = df_new['Title'].str.replace('\t','')
df_new.head()
```

# Plotting Datasets

```python
title_count = df_new.pivot_table(index='Year', 
                                  columns='Month', 
                                  values='Title', aggfunc='count')

title_count['Total'] = title_count.sum(axis=1)
title_count['Mean'] = np.round(title_count['Total'] / 12)
title_count
```

```python
fig = plt.figure(figsize=(9,4))
title_count['Total'].plot(color='red', marker='.', linewidth=1.5)
plt.title('Total Journal PTIIK Publications in 2017 - 2022', fontsize=20, pad=20, color='orange')
plt.xlabel('Journal Publication Year', fontsize=15, color='blue')
plt.ylabel('Journal Totals', fontsize=15, color='blue')
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(300, 800)
plt.show()
```

```python
highest_jp_year = title_count['Total'].idxmax()
highest_jp_year
```

```python
df_highest_jp_year = df_new.groupby('Year').get_group(highest_jp_year)
df_highest_jp_year
```

```python
df_highest_jp_year.groupby('Month')['Title'].count().plot(marker='.')
plt.xticks(np.arange(1, 13, 1))
plt.show()
```
